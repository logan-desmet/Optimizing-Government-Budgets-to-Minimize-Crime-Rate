This takes the budget data and pivots the columns to the rows, and constructs the proper budget categories. Also converts all datapoints to proportions for ML.

```python
import pandas as pd
```

```python
budgetData = pd.read_csv("budgetCombinedFinal.csv")
```

```python
budgetData = budgetData.fillna(0)
```

```python
# add a total column
totalCol = budgetData.iloc[:, 2:].sum(axis=1)

budgetData["Total"] = totalCol
```

```python
budgetData["Total"]
```

```python
colToProp = budgetData.columns[2:-1]
budgetData[colToProp] = budgetData[colToProp].div(budgetData["Total"], axis=0)
```

Now budgets are the proportion of total budget:

```python
# output to csv
budgetData.to_csv("budgetCombinedPostPrep_FINAL.csv")
```
