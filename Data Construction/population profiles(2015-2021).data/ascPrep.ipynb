```python
import pandas as pd
```

```python
acsdata = pd.read_csv("ACSSPP1Y2015-Data.csv")
#acsdata
```

```python
# columns to list
tempcolList = []
for col in acsdata:
  colname = acsdata[col][0]
  #print(colname)
  try:
    if colname.startswith("Margin"):
      acsdata = acsdata.drop([col], axis = 1)
    else:
      tempcolList.append(colname)
  except:
    acsdata = acsdata.drop([col], axis = 1)
```

```python
newColList = []

for col in tempcolList:
  if col.startswith("Estimate"):
    temp = col.replace("Estimate!!", "")
    newColList.append(temp)
  else:
    newColList.append(col)
```

```python
noparenColList = []

for word in newColList:
  if "(IN" in word:
    temp = word.split("(IN")
    fparttemp = temp[0]
    bparttemp = temp[1].split("!!")
    finalcol = "(" + fparttemp + ") "
    for w in bparttemp[1:]:
      finalcol += w + "_"

    noparenColList.append(finalcol[:-1])
  else:
    noparenColList.append(word)
```

```python
colList = []

for col in noparenColList:
  colSplit = col.split("!!")
  if len(colSplit) == 1:
    colList.append(colSplit[0])
  else:
    tempstring = "(" + colSplit[0] + ") "
    for word in colSplit[1:]:
      tempstring += word+"_"
    colList.append(tempstring[:-1]) #remove trailing "_"
```

```python
# make colList the col names and remove that row from df

#acsdata.columns = colList
#acsdata = acsdata.drop([0], axis=0)

# replace row 0 instead with new col names
acsdata.loc[0] = colList
```

```python
#output to df
acsdata.to_csv("acs2015_postPrep.csv")
```
